```kotlin
fun main() {
    val spices = listOf("curry", "pepper", "cayenne", "ginger", "red curry", "green curry", "red pepper")

    // 1️⃣ Todos los curries ordenados por longitud
    val curries = spices.filter { it.contains("curry") }.sortedBy { it.length }
    println("Curries: $curries")

    // 2️⃣ Spices que empiezan con 'c' y terminan con 'e' (dos formas)
    val cAndE1 = spices.filter { it.startsWith('c') && it.endsWith('e') }
    val cAndE2 = spices.filter { it.first() == 'c' && it.last() == 'e' }
    println("C and E (1): $cAndE1")
    println("C and E (2): $cAndE2")

    // 3️⃣ Primeras tres que empiezan con 'c'
    val firstThreeC = spices.take(3).filter { it.startsWith('c') }
    println("First three starting with c: $firstThreeC")
}
```
