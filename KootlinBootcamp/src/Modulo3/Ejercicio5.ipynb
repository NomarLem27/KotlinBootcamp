```kotlin
fun main() {
    println(whatShouldIDoToday("happy"))
    println(whatShouldIDoToday("sad", "rainy", 10))
    println(whatShouldIDoToday("energetic", "sunny", 30))
}

fun whatShouldIDoToday(mood: String, weather: String = "sunny", temperature: Int = 24): String {
    return when {
        mood == "happy" && weather == "sunny" -> "Go for a walk!"
        mood == "sad" && weather == "rainy" -> "Stay home and read."
        mood == "energetic" && temperature > 25 -> "Go play some sport outside!"
        else -> "Stay home and rest."
    }
}
```
