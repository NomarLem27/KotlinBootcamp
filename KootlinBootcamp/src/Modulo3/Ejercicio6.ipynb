```kotlin
fun main() {
    print("How are you feeling today? ")
    val mood = readLine()!!
    println(whatShouldIDoToday(mood, "rainy", 0))
}

fun whatShouldIDoToday(mood: String, weather: String = "sunny", temperature: Int = 24): String {
    fun isVeryHot() = temperature > 35
    fun isSadAndCold() = mood == "sad" && weather == "rainy" && temperature == 0
    fun isHappyAndSunny() = mood == "happy" && weather == "sunny"

    return when {
        isVeryHot() -> "Go swimming!"
        isSadAndCold() -> "Stay in bed."
        isHappyAndSunny() -> "Go for a walk!"
        mood == "bored" -> "Watch a movie!"
        mood == "angry" -> "Do some exercise to relax!"
        else -> "Stay home and read."
    }
}
```
