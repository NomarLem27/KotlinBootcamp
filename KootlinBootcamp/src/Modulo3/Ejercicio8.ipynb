```kotlin
fun main() {
    val rollDice = { (1..12).random() }
    println("Dice roll: ${rollDice()}")

    val rollDiceSides = { sides: Int ->
        if (sides == 0) 0 else (1..sides).random()
    }
    println("Roll 6 sides: ${rollDiceSides(6)}")

    val rollDice2: (Int) -> Int = { sides ->
        if (sides == 0) 0 else (1..sides).random()
    }
    println("Roll with rollDice2: ${rollDice2(8)}")
}
```
