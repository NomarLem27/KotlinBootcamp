```kotlin

class Book(val title: String, val author: String, val year: Int) {
    fun getTitleAuthor(): Pair<String, String> = Pair(title, author)
    fun getTitleAuthorYear(): Triple<String, String, Int> = Triple(title, author, year)
}

fun main() {
    val book = Book("The Hobbit", "J.R.R. Tolkien", 1937)
    val (title, author) = book.getTitleAuthor()
    val (t, a, y) = book.getTitleAuthorYear()
    println("Here is your book $t written by $a in $y.")
}
```
