```kotlin
val MAX_BOOKS = 5

class Book(val title: String) {
    fun canBorrow(currentlyBorrowed: Int): Boolean {
        return currentlyBorrowed < MAX_BOOKS
    }

    companion object Constants {
        const val BASE_URL = "https://library.com/books/"
    }

    fun printUrl() {
        println("${Constants.BASE_URL}${title.replace(" ", "_")}.html")
    }
}

fun main() {
    val book = Book("Moby Dick")
    println(book.canBorrow(3))
    book.printUrl()
}
```
