```kotlin
fun main() {
    val allBooks = setOf("Hamlet", "Othello", "Macbeth", "King Lear")
    val library = mapOf("William Shakespeare" to allBooks)
    println(library.any { it.value.contains("Hamlet") })

    val moreBooks = mutableMapOf("Romeo and Juliet" to "William Shakespeare")
    moreBooks.getOrPut("Hamlet") { "William Shakespeare" }
    moreBooks.getOrPut("The Tempest") { "William Shakespeare" }
    println(moreBooks)
}
```
