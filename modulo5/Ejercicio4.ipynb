```kotlin
import kotlin.random.Random

class Book(var pages: Int)

fun Book.weight(): Double = pages * 1.5

fun Book.tornPages(torn: Int) {
    pages = if (pages - torn < 0) 0 else pages - torn
}

class Puppy {
    fun playWithBook(book: Book) {
        val torn = Random.nextInt(1, 10)
        book.tornPages(torn)
        println("Puppy tore $torn pages! Remaining: ${book.pages}")
    }
}

fun main() {
    val book = Book(50)
    val puppy = Puppy()
    while (book.pages > 0) {
        puppy.playWithBook(book)
    }
    println("No more pages left!")
}
```
