```kotlin

fun main() {
    val spice = SimpleSpice()
    println("Spice: ${spice.name}, Heat: ${spice.heat}")
}


class SimpleSpice {

    var name: String = "curry"
    var spiciness: String = "mild"


    val heat: Int
        get() = when (spiciness.lowercase()) {
            "mild" -> 5
            "medium" -> 7
            "spicy" -> 9
            "very spicy" -> 10
            else -> 0
        }
}
```
